```python
thetas = data.iloc[:, 0]
thetas_h001 = data_h001.iloc[:, 0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
mean = statistics.mean(thetas)
sd = statistics.stdev(thetas)
  
plt.plot(thetas, norm.pdf(thetas, mean, sd))
```
